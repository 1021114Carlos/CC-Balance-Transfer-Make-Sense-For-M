```python
import matplotlib.pyplot as plt
from matplotlib_venn import venn2
```

```python
# Define the labels for the sets
labels = {'10': 'Brain and Cognitive Science Majors\nOnly',
          '01': 'Sophomores\nOnly',
          '11': 'Both'}

# Create the Venn diagram
plt.figure(figsize=(8, 8))
venn = venn2(subsets=(1, 1, 1), set_labels=('Brain and Cognitive Science Majors', 'Sophomores'))
venn.get_label_by_id('10').set_text(labels['10'])
venn.get_label_by_id('01').set_text(labels['01'])
venn.get_label_by_id('11').set_text(labels['11'])

# Shade the entire Venn diagram to illustrate the first law
for subset in ('10', '01', '11'):
    venn.get_patch_by_id(subset).set_color('lightblue')
    venn.get_patch_by_id(subset).set_alpha(0.5)

# Set the title
plt.title('Venn Diagram for Brain and Cognitive Science Majors and Sophomores\nIllustrating the First Law')

# Show the plot
plt.show()
```
